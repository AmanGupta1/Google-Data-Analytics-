```python
from glob import glob
import pandas as pd
import csv
from pandas import read_csv
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv('/home/ubuntu/merged_files/kmeans_task_events.csv')
```

```python
df1
```

```python
df2=df1
df2.drop(df1.columns[[5,10]], axis=1, inplace=True)
```

```python
df2
```

```python
kmeans=KMeans(n_clusters=6)
kmeans.fit(df2)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
{i: np.where(kmeans.labels_ == i)[0] for i in range(kmeans.n_clusters)}
```

```python
df1['clusters']=kmeans.labels_
```

```python
df1
```

```python
df2 = df1.groupby(['clusters', 'event type']).agg({'job ID': 'count'})
```
